```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

I=pd.read_csv(r"C:\Users\loveo\OneDrive\Desktop\Coursera\Machine Learning\Regression\insurance.csv")

X=I.iloc[:, :-1].values
y=I.iloc[:,-1].values

print(X)
y = y.reshape(len(y),1)
print(y)
```

Data preprocessing

Feature scaling

```python
from sklearn.preprocessing import StandardScaler
sc_X= StandardScaler()
X[:,[0,2]] = sc_X.fit_transform(X[:,[0,2]] )

print(X)
print(y.shape)
```

```python
#encoding
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

# Specify the columns to one-hot encode
columns = [1, 4, 5]  # Adjust these indices based on your dataset

# Create a ColumnTransformer with OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), columns)], remainder='passthrough')

# Apply the ColumnTransformer to the data
X = np.array(ct.fit_transform(X))

print(X)
print(X.shape)
print(y.shape)
```

Data split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state =0)
print (X_train.shape)
```

```python
#Training the multiple linear regression model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)

#predicting 
y_pred = (regressor.predict(X_test))


np.set_printoptions(precision=5)

print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))

print(regressor.predict(([[0,1,1,0,0,0,1,0,-0.7979535482,0.3833068472,3]])))
```

```python

from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)
print(r2)
```

```python
# Polynomial regression model
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
poly_reg = PolynomialFeatures(degree = 2)
X_poly = poly_reg.fit_transform(X_train)
regressor = LinearRegression()
regressor.fit(X_poly, y_train)

y_pred = regressor.predict(poly_reg.transform(X_test))
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))

from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)
print(r2)
print(regressor.predict(poly_reg.transform([[0,1,1,0,0,0,1,0,-0.7979535482,0.3833068472,3]])))
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 0)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)
print(r2)

print(regressor.predict(([[0,1,1,0,0,0,1,0,-0.7979535482,0.3833068472,3]])))
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor.fit(X_train,y_train)

y_pred = regressor.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)
print(r2)

print(regressor.predict(([[0,1,1,0,0,0,1,0,-0.7979535482,0.3833068472,3]])))
```
